# Class :

## A class is a code template for creating objects. Objects have member variables and have behaviour associated with them.In python a class is created by the keyword class. An object is created using the constructor of the class. This object will then be called the instance of the class.

# Object :

## Python is an object-oriented programming language, This defines a set of attributes that will characterize any object that is instantiated from this class. Object â€” An instance of a class. This is the realized version of the class, where the class is manifested in the program.

# Attribute :

## An instance attribute is a Python variable belonging to one and only one object. A class attribute is a Python variable that belongs to a class rather than a particular object. It is shared between all the objects of this class and it is defined outside the constructor function of the class.

# Behaviour :

## Objects in Python are generally classified according to their behaviors and the features that they implement. For example, all of the sequence types such as strings, lists, and tuples are grouped together merely because they all happen to support a common set of sequence operations

```python
class Car():
    def __init__(self,name,model,colour,doors,fuel):
        self.name = name
        self.model = model
        self.colour = colour
        self.doors = doors
        self.fuel = fuel

    def average(self,kilo):
        print("Average :",kilo,"Kilometers Per Liter")

    def speed(self,kilo):
        print("Speed :",kilo,"Kilometers Per Hour")
        
    def nitro(self,kilo):
        print("Nitro :",kilo,"Meters Per sec")

car_1 = Car("Mercedes-Benz","Maybach","White","4","30")
car_2 = Car("BMW","M8 Gran Coupe","Grey","4","25")
car_3 = Car("Ferrari","Tributo","Red","2","20")
car_4 = Car("Mazda","MX-5","Green","2","15")
car_5 = Car("Audi","Q8","Black","4","10")

def car_10():
    print("Name :",car_1.name)
    print("Model :",car_1.model)
    print("Colour :",car_1.colour)
    car_1.average(30)
    car_1.speed(30)
    car_1.nitro(30)

def car_20():
    print("Name :",car_2.name)
    print("Model :",car_2.model)
    print("Colour :",car_2.colour)
    car_1.average(25)
    car_1.speed(25)
    car_1.nitro(25)    

def car_30():
    print("Name :",car_3.name)
    print("Model :",car_3.model)
    print("Colour :",car_3.colour)
    car_1.average(20)
    car_1.speed(20)
    car_1.nitro(20)

def car_40():
    print("Name :",car_4.name)
    print("Model :",car_4.model)
    print("Colour :",car_4.colour)
    car_1.average(15)
    car_1.speed(15)
    car_1.nitro(15)

def car_50():
    print("Name :",car_5.name)
    print("Model :",car_5.model)
    print("Colour :",car_5.colour)
    car_1.average(10)
    car_1.speed(10)
    car_1.nitro(10)

car_10()
print("--------------------------------------")
car_20()
print("--------------------------------------")
car_30()
print("--------------------------------------")
car_40()
print("--------------------------------------")
car_50()
```
